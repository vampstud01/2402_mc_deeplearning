# 퍼셉트론의 한계

```python
class 퍼셉트론:
    def __init__(self):
        self.w = None
        self.b = None

    def __call__(self, x):
        z = np.dot(x, self.w) + self.b
        y = np.where(z > 0, 1, -1)
        return y

    def 학습(self, Xs, ys, 학습횟수, 학습률=1.0):
        assert len(Xs) == len(ys)
        # 매개변수 초기화
        s, n = Xs.shape
        self.w = np.zeros(n, dtype='float32')
        self.b = 0.0

        for 에폭 in range(학습횟수):
            print(f'[학습 {에폭}] w={self.w}, b={self.b}')
            for xi, yi in zip(Xs, ys):
                output = self(xi)
                error = yi - output
                갱신 = error * 학습률
                self.w += 갱신 * xi
                self.b += 갱신
```

```python
Xs = np.array([(0, 0), (0, 1), (1, 0), (1, 1)])
y_or = np.array([0, 1, 1, 1])
y_xor = np.array([0, 1, 1, 0])

z0 = lambda x1, w, b: (w[0] * x1 + b) / -w[1]
x1 = np.arange(-1.0, 2.0, 0.1)

plt.figure(figsize=(4, 2))
# OR
plt.subplot(1, 2, 1)
plt.scatter(Xs[:, 0], Xs[:, 1], c=y_or)
plt.plot(x1, z0(x1, w=np.array([1.0, 1.0]), b=-0.5))
# 보조선
plt.hlines(y=0, xmin=-1.0, xmax=2.0, linestyle='--', color='black')
plt.vlines(x=0, ymin=-1.5, ymax=1.5, linestyle='--', color='black')
# XOR
plt.subplot(1, 2, 2)
plt.scatter(Xs[:, 0], Xs[:, 1], c=y_xor)
plt.plot(x1, z0(x1, w=np.array([-0.2, -0.2]), b=0.2))
# 보조선
plt.hlines(y=0, xmin=-1.0, xmax=2.0, linestyle='--', color='black')
plt.vlines(x=0, ymin=-1.5, ymax=1.5, linestyle='--', color='black')
plt.show()
```

```python
model = 퍼셉트론()
model.학습(Xs, ys=np.array([-1, 1, 1, -1]), 학습횟수=5, 학습률=0.1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

random = np.random.RandomState(123)
x = np.linspace(-5, 5, 100)
noise = random.randn(len(x))
y = x ** 2 + noise

Xs = x.reshape(-1, 1)
X2 = np.hstack([Xs, Xs ** 2])
model = LinearRegression()
model.fit(X2, y)
print(f'R2={r2_score(y, model.predict(X2)):.3f} < {r2_score(y, y)}')

plt.scatter(x, y)
plt.plot(x, y - noise, 'r--')
plt.plot(x, model.predict(X2), 'k')
plt.show()
```

# 다층 퍼셉트론

Multi Layer Perceptron (MLP)

```python
AND = 퍼셉트론()
AND.w = np.array([0.5, 0.5])
AND.b = -0.7
NAND = 퍼셉트론()
NAND.w = np.array([-0.5, -0.5])
NAND.b = 0.7
OR = 퍼셉트론()
OR.w = np.array([0.5, 0.5])
OR.b = -0.2

Xs = np.array([(0, 0), (0, 1), (1, 0), (1, 1)])
assert all(AND(Xs) == [-1, -1, -1, 1])
assert all(NAND(Xs) == [1, 1, 1, -1])
assert all(OR(Xs) == [-1, 1, 1, 1])

s1 = NAND(Xs)
s2 = OR(Xs)
S = np.array([s1, s2]).T
assert all(AND(S) == [-1, 1, 1, -1])
```

## 활성화 함수

1. 출력
2. 변환

```python
def step(z, ymin=-1, ymax=1):
    return np.where(z > 0, ymax, ymin)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.arange(-5, 5, 0.1)
plt.figure(figsize=(2, 2))
plt.plot(z, step(z, ymin=0))
plt.plot(z, sigmoid(z))
plt.plot(z, step(z))
plt.plot(z, np.tanh(z))
plt.show()
```

## 순전파

```python
xi = np.array([1.0, 0.5])
```

```python
W1 = np.random.randn(2, 3)
b1 = np.random.randn(3)
z1 = np.dot(xi, W1) + b1
z1
```

```python
W2 = np.random.randn(3, 2)
b2 = np.random.randn(2)
z2 = np.dot(z1, W2) + b2
z2
```

```python
W3 = np.random.randn(2, 2)
b3 = np.random.randn(2)
z3 = np.dot(z2, W3) + b3
z3
```

```python
class 완전연결:
    def __init__(self, 입력수, 출력수, 활성화=None):
        self.W = np.random.randn(입력수, 출력수)
        self.b = np.random.randn(출력수)
        self.activation = 활성화

    def __call__(self, Xs):
        zs = np.dot(Xs, self.W) + self.b
        if self.activation:
            return self.activation(zs)
        return zs # 항등 함수

class 신경망:
    def __init__(self):
        self.layers = []

    def add(self, 계층):
        self.layers.append(계층)

    def __call__(self, Xs):
        """순전파 (feed forward)"""
        출력 = Xs
        for 계층 in self.layers:
            출력 = 계층(출력)
        return 출력

FC = Dense = Affine = Linear = 완전연결
```

```python
layer1 = FC(2, 3, sigmoid)
layer2 = FC(3, 2, sigmoid)
layer3 = FC(2, 2)

# (수동)순전파
h1 = layer1(xi)
h2 = layer2(h1)
z3 = layer3(h2)
print(f'x={xi} \n-> h1={h1}')
print(f'h1={h1} \n-> h2={h2}')
print(f'h2={h2} \n-> z3={z3}')

model = 신경망()
model.add(layer1)
model.add(layer2)
model.add(layer3)
assert all(model(xi) == z3)
```

이번 실습의 목표는 은닉층에서 비선형 변환이 있어야 신경망 계층이 효력을 발생한다는 점을 시각적으로 확인하는데 있습니다.

```python
Xs = np.array([(0, 0), (0, 1), (1, 0), (1, 1)])
y_xor = np.array([0, 1, 1, 0])

layer1 = FC(2, 2)
Zs = layer1(Xs)

plt.figure(figsize=(4, 2))
plt.subplot(1, 2, 1)
plt.scatter(Xs[:, 0], Xs[:, 1], c=y_xor)
plt.subplot(1, 2, 2)
plt.scatter(Zs[:, 0], Zs[:, 1], c=y_xor)
plt.show()
```
